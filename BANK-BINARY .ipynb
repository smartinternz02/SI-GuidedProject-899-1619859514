```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc =StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
from tensorflow.keras.models import Sequential #-used to initialize the model.
from tensorflow.keras.layers import Dense #- used to add layers of ANN
```

```python
b= Sequential()
```

```python
b.add(Dense(12,kernel_initializer="random_uniform",activation='relu'))
```

```python
b.add(Dense(24,kernel_initializer="random_uniform",activation='relu'))
```

```python
b.add(Dense(1,kernel_initializer="random_uniform",activation='sigmoid'))
```

```python
b.compile("adam","binary_crossentropy", metrics= ["accuracy"])
```

```python
b.fit(x_train,y_train,batch_size=32,epochs =150)
```
